```python
# 1.
import numpy as np
import pandas as pd

data = [3, 2, 5, 4, 7, 2, 3, 3, 1, 6, 4, 2, 3, 5, 2, 4, 2, 1, 3, 5,
        6, 3, 2, 1, 4, 2, 4, 5, 3, 2, 7, 2, 3, 4, 5, 1, 6, 2, 4, 3,
        5, 3, 2, 4, 2, 6, 3, 2, 4, 5]

s = pd.Series(data)

mean = np.mean(s)
median = np.median(s)
mode = s.mode()[0]

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
```

```python
# 2 
data = [3, 5, 2, 4, 6, 2, 3, 4, 2, 5,
        7, 2, 3, 4, 2, 4, 2, 3, 5, 6,
        3, 2, 1, 4, 2, 4, 5, 3, 2, 7,
        2, 3, 4, 5, 1, 6, 2, 4, 3, 5,
        3, 2, 4, 2, 6, 3, 2, 4, 5, 3]

df = pd.Series(data)

range_v = np.max(df) - np.min(df)
variance = np.var(df)
std_d = np.std(df)

print("Range : ",range_v)
print("Variance : ",variance)
print("Standard Deviation : ",std_d)
```
