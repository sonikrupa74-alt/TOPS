```python
import numpy as np
import pandas as pd

data = [3, 2, 5, 4, 7, 2, 3, 3, 1, 6, 4, 2, 3, 5, 2, 4, 2, 1, 3, 5,
        6, 3, 2, 1, 4, 2, 4, 5, 3, 2, 7, 2, 3, 4, 5, 1, 6, 2, 4, 3,
        5, 3, 2, 4, 2, 6, 3, 2, 4, 5]

s = pd.Series(data)

mean = np.mean(s)

median = np.median(s)

mode = s.mode()[0]

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
```
